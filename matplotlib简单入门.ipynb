## 两种绘图方式的区分

### 函数式绘图

```python
plt.figure(figsize=(10, 6))
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.plot(x, y, 'g--', lw=2, label='sinx')

# 调整坐标轴范围
x_min = 0
y_min = -1.5
x_max = 10
y_max = 1.5

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# 设置坐标轴标签
plt.xlabel('x轴', fontsize=15)
plt.ylabel('y轴', fontsize=15)

# 设置坐标轴标签
x_location = np.arange(0, 10 ,2)
x_labels = ['2019-01-01', '2019-02-01', '2019-03-01', '2019-04-01', '2019-05-01']
y_location = np.arange(-1, 1.5, 1)
y_labels = [u'最小值', u'零值',u'最大值']
plt.xticks(x_location, x_labels, rotation=45, fontsize=15)
plt.yticks(y_location, y_labels)

# 设置网格
plt.grid(True, ls=':', color='r', alpha=0.5)

# 设置标题
plt.title(u"函数式绘图 vs 对象式绘图", fontsize=25)

# 添加图例
plt.legend(loc="upper right", fontsize=15)
plt.show;
```

### 标注点的绘制

当我们需要给数据添加指向性注释文本时，可以使用matplotlib的annotate（）函数，他支持箭头指示
- s：注释文本内容
- xy: 被注释的的坐标点，二维元组模式（x,y)
- xytext：注释文本的坐标点
- xycoords：被注释点的坐标系属性，默认为’data‘
- textcoords：设置注释文本的坐标系属性，默认与xycoords属性值相同，通常设置为’offset point‘或者’offset pixels‘，即相对于被注释点的偏移量
- arrowprops：设置箭头的样式，字典（dict）格式
- bbox： 设置文本周围所添加的外框属性

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
x = np.arange(10,20)
y = np.around(np.log(x), 2)
ax.plot(x,y,marker='o')
ax.annotate(u'样式1', xy=(x[1],y[1]), xytext=(80, 10), textcoords='offset points', 
            arrowprops=dict(arrowstyle='->', connectionstyle='angle3,angleA=80,angleB=50'))
ax.annotate(u'样式2', xy=(x[3],y[3]), xytext=(80, 10), textcoords='offset points', 
            arrowprops=dict(facecolor='black', shrink=0.05, width=5))
ax.annotate(u'样式3', xy=(x[5],y[5]), xytext=(80, 10), textcoords='offset points', 
            arrowprops=dict(facecolor='green',headwidth=5, headlength=10),
            bbox=dict(boxstyle='circle,pad=0.5', fc='yellow', ec='k', lw=1, alpha=0.5))
#ec: edgecolor
ax.annotate(u'样式4', xy=(x[7],y[7]), xytext=(80, 10), textcoords='offset points', 
           arrowprops=dict(facecolor='blue',headwidth=5, headlength=10),
           bbox=dict(boxstyle='round,pad=0.5', fc='gray', ec='k', lw=1, alpha=0.5));
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y, 'g--', lw=2, label='sinx')

# 调整坐标轴范围
x_min = 0
y_min = -1.5
x_max = 10
y_max = 1.5

ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

# 设置坐标轴标签
x_location = np.arange(0, 10 ,2)
x_labels = ['2019-01-01', '2019-02-01', '2019-03-01', '2019-04-01', '2019-05-01']
y_location = np.arange(-1, 1.5, 1)
y_labels = [u'最小值', u'零值', u'最大值']
ax.set_xticks(x_location)
ax.set_yticks(y_location)
ax.set_xticklabels(x_labels, rotation=45, fontsize=15)
ax.set_yticklabels(y_labels, fontsize=15)

# 设置轴标签
ax.set_xlabel('x轴',fontsize=15)
ax.set_ylabel('y轴',fontsize=15)

# 设置网格
ax.grid(True, ls=':', color='r', alpha=0.5)

# 设置标题
ax.set_title(u"函数式绘图 vs 对象式绘图", fontsize=25)

# 添加sin（）的最高点注释
ax.annotate(u'最高点',
            xy = (np.pi/2, 1),
            xytext = (np.pi/2, 1.3), #注释文本左端的坐标点
            weight = 'regular', # 注释文本的字体风格， bold为粗体
            color = 'g',
            fontsize = 15,
            arrowprops = {'arrowstyle': '->',
                         'connectionstyle': 'arc3',
                         'color': 'g'}
           )

# 添加sin（）的最低点注释
ax.annotate(u'最低点',
            xy = (np.pi*3/2, -1),
            xytext = (np.pi*3/2, -1.3), # 注释文本左端的坐标点
            weight = 'regular',  #注释文本的字体风格， bold为粗体
            color = 'g',
            fontsize = 15,
            arrowprops = {'arrowstyle': '->',
                         'connectionstyle': 'arc3',
                         'color': 'r'}
           )            

# 添加图例
ax.legend(loc="upper right", fontsize=15)
plt.show;
```

### 直方图的绘制

- x: 绘制直方图的数据
- bins: 直方图柱数
- density：是否将频数转换为频率
- n：直方图每一个bar区间数据的频数或者频率
- bins：用于返回各个bin的区间范围
- patches：列表返回每个bin的图形对象

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

ax.hist(np.random.normal(loc=0, scale=1, size=1000), bins=50, density=False, color='b')
ax.set_xlabel(u'样本值', fontsize=15)
ax.set_ylabel(u'频数', fontsize=15)
ax.set_title(u'正态分布直方图', fontsize=25);
```
